```python
from jobspy import scrape_jobs
import pandas as pd
from IPython.display import display, HTML
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 50)
```

```python
# example 1 (no hyperlinks, USA)
result = scrape_jobs(
    site_name=["linkedin", "zip_recruiter"],
    location='san francisco',
    search_term="engineer",
    results_wanted=5, 
)

display(result.jobs)
display(result.errors)
```

```python
# example 3 - remote USA & hyperlinks
result = scrape_jobs(
    site_name=["linkedin", "zip_recruiter", "indeed"],
    # location='san francisco',
    search_term="software engineer",
    country_indeed="USA",
    hyperlinks=True,
    is_remote=True,
    results_wanted=5, 
)
```

```python
# use if hyperlinks=True
html = result.jobs.to_html(escape=False)
# change max-width: 200px to show more or less of the content
truncate_width = f'<style>.dataframe td {{ max-width: 200px; overflow: hidden; text-overflow: ellipsis; white-space: nowrap; }}</style>{html}'
display(HTML(truncate_width))
display(result.errors)
```

```python
# example 3 - with hyperlinks, international - linkedin (no zip_recruiter)
result = scrape_jobs(
    site_name=["linkedin"],
    location='berlin',
    search_term="engineer",
    hyperlinks=True,
    results_wanted=5,
    easy_apply=True
)
```

```python
# use if hyperlinks=True
html = result.jobs.to_html(escape=False)
# change max-width: 200px to show more or less of the content
truncate_width = f'<style>.dataframe td {{ max-width: 200px; overflow: hidden; text-overflow: ellipsis; white-space: nowrap; }}</style>{html}'
display(HTML(truncate_width))
display(result.errors)
```
